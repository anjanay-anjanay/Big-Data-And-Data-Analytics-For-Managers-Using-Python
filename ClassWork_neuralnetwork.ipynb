```python
# 1.0 Call libraries
import pandas as pd
import numpy as np
from  pathlib import Path
```

```python
# 1.0.1
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# 2.1 Where is our data file:
pathToFolder = "/gdrive/MyDrive/deeplearning"
```

```python
# 2.2 Full file path:
path = Path(pathToFolder) / "mobile_price.csv"
```

```python
# 3.0 Read data
mobile_price = pd.read_csv(path)
mobile_price.head()
```

```python
mobile_price.tail()
```

```python
# 3.1
mobile_price.shape   # (2000, 21)
mobile_price.dtypes
```

```python
# 3.2 Get target and predictors:
y = mobile_price.pop("ram")
X = mobile_price #Convention: predictor in Capital X
```

```python
X.columns
```

```python
# 4.0 Build model:

model = tf.keras.Sequential()
```

```python
# 4.1
model.add(layers.Input( shape = (X.shape[1],) ) )
#input layer #11 neuron
```

```python
model.add(layers.Dense(11,) )
```

```python
model.add(layers.Dense(11,activation="tanh"))
```

```python
model.add(layers.Dense(770,activation="relu"))
```

```python
model.add(layers.Dense(770,activation="sigmoid"))
```

```python
#4.3 Compile
model.compile( loss = "mae")
#mean absolute error
```

```python
# 4.4 Train
model.fit(X,y, epochs = 128 )
```

```python
from tensorflow.keras.utils import plot_model
```

```python
plot_model(model,show_shapes=True)
```
